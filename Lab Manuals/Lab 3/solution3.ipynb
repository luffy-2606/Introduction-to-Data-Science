```python
def add(num1, num2):
  try:
    return num1+num2
  except TypeError as e:
    print("Cannot add the variables. The exception was:", e)
```

```python
def sub(num1, num2):
  try:
    return num1-num2
  except TypeError as e:
    print("Cannot subtract the variables. The exception was:", e)
```

```python
def mult(num1, num2):
  try:
    return num1*num2
  except TypeError as e:
    print("Cannot multiply the variables. The exception was:", e)
```

```python
def div(num1, num2):
  try:
    return num1/num2
  except TypeError as e:
    print("Cannot divide the variables. The exception was:", e)
  except ZeroDivisionError as e:
    print("Cannot divide the variables. The exception was:", e)
```

```python
def sum_file(p1, p2):
  try:
    with open(p1, "r") as file:
      FileLines = file.readlines()
  except FileNotFoundError:
    print("The file to read was not found!")
  try:
    total = 0
    for j in range(0,len(FileLines)):
      total = total + int(FileLines[j])
  except ValueError:
    print("Incorrect type in file!")

  try:
    with open(p2, "w") as file:
      file.write(str(total))
  except FileNotFoundError:
    print("The file to read was not found!")
  except PermissionError:
    print("Permission Error!")

m1='/content/p1.txt'
m2='/content/p2.txt'
sum_file(m1, m2)
```

# Q4


```python
class Book:
  def __init__(self, title, author, year):
    self.title = title
    self.author = author
    self.year = year
    self.available=True

  def get_info(self):
    if (self.available == True):
      return self.title, "(", self.year, ") by ", self.author, " - Available"
    else:
      return self.title, "(", self.year, ") by ", self.author, " - Checked Out"

  def checkout(self):
    if (self.available == True):
      self.available = False
      print("You have checked out ", self.title)
    else:
      print("Sorry, ", self.title, " is not available")

  def return_book(self):
    if (self.available == True):
      print("This book was not checked out")
    else:
      self.available = True
      print("You have returned ", self.title)


book1 = Book("The Hobbit", "J.R.R. Tolkien", 1937)
print(book1.get_info()) # "The Hobbit (1937) by J.R.R. Tolkien - Available"
book1.checkout() # "You have checked out 'The Hobbit'."
print(book1.get_info()) # "The Hobbit (1937) by J.R.R. Tolkien - Checked Out"
book1.checkout() # "Sorry, 'The Hobbit' is currently not available."
book1.return_book() # "You have returned 'The Hobbit'."
book1.return_book() # "This book was not checked out."
```
