# 4. Use the given arrays to answer the questions below:
a = np.array([1, 2, 3])\
b = np.array([4, 5, 6])\
a. Add a and b.\
b. Multiply a and b element-wise.\
c. Calculate the dot product of a and b.\
d. Find the mean and standard deviation of b.


```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

add1 = a+b
print("Addition:",add1)

mult1 = a*b
print("Multiplication:",mult1)

dot1 = np.dot(a, b)
print("Dot Product",dot1)

mean1 = np.mean(b)
std1 = np.std(b)

print("Mean:", mean1)
print("Standard Deviation:", std1)
```

# 5. Given two numpy arrays A and B, apply the formula given below

```python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])
z = (A*A)+np.sin(B)
sub = A-B
base = np.abs(sub)
base = np.sqrt(base+1)
z = z / base
print(z)
```

# 6. Run the code given below and explain its output in 2-3 lines.

```python
import numpy as np
arr = np.random.randint(-50, 151, size=100)
arr_reshaped = arr.reshape((10, 10))
arr_reshaped[arr_reshaped < 0] = 0
arr_min = arr_reshaped.min()
arr_max = arr_reshaped.max()
normalized = (arr_reshaped - arr_min) / (arr_max - arr_min)
row_means = normalized.mean(axis=1)
print("Original reshaped array (negatives replaced):\n", arr_reshaped)
print("\nNormalized array:\n", normalized)
print("\nRow-wise means:\n", row_means)


# Explanation:
# the reshaped array has a range of positive values from 0 to 150
# the normalized array has
# The mean of the normalized array rows are calclated resulting in a 10 number array
```
