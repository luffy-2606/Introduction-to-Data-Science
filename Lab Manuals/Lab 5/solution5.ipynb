```python
import requests
from bs4 import BeautifulSoup

URL = "https://lhr.nu.edu.pk/faculty/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

departments = {
    "fsc": "FAST School of Computing",
    "ee": "Electrical Engineering",
    "ce": "Civil Engineering",
    "fms": "FAST School of Management",
    "sh": "Sciences & Humanities"
}

for dept_id, dept_name in departments.items():
    results = soup.find(id=dept_id)
    if not results:
        continue

    faculty = results.find_all("div", class_="col-lg-3 col-md-4 col-sm-6 col-12")
    for card in faculty:
        link = card.find("a")["href"]
        emp_id = link[-4:]
        profile_url = "https://lhr.nu.edu.pk" + link

        name = card.find("h5", class_="text-center").text.strip()

        info_lines = card.find("p", class_="small text-center font-italic").text.strip().split("\n")
        designation = info_lines[0].strip()
        extra = [line.strip() for line in info_lines[1:] if line.strip()]

        email = card.find("p", class_="mb-0 text-center").text.strip()

        image = card.find("img").get("src")
        if image.startswith("/"):
            image = "https://lhr.nu.edu.pk" + image
        extension = ""
        try:
            prof_page = requests.get(profile_url)
            soup2 = BeautifulSoup(prof_page.content, "html.parser")
            contact_div = soup2.find("div", class_="col-md-6")
            if contact_div:
                contact_text = contact_div.get_text(" ", strip=True)
                if "Ext:" in contact_text:
                    extension = contact_text.split("Ext:")[-1].strip().split()[0]
        except:
            extension = ""

        print("Employee ID:", emp_id)
        print("Name:", name)
        print("Designation:", designation)
        print("Extra Info:", extra)
        print("Email:", email)
        print("Extension:", extension)
        print("Image URL:", image)
        print("Department:", dept_name)
        print("Profile URL:", profile_url)
        print("\n")
```
